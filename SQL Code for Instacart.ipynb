# Creating orders_products tables

Select o.order_id, op.product_id, op.add_to_cart_order, op.reordered, o.user_id, o.order_number, o.order_dow, o.order_hour_of_day, o.days_since_prior_order into order_products_prior_2
from order_products_prior op join orders o on o.order_id = op.order_id
group by o.order_id, op.product_id, op.add_to_cart_order, op.reordered, o.user_id, o.order_number, o.order_dow, o.order_hour_of_day, o.days_since_prior_order;


Select o.order_id, op.product_id, op.add_to_cart_order, op.reordered, o.user_id, o.order_number, o.order_dow, o.order_hour_of_day, o.days_since_prior_order into order_products_train_2
from order_products_train op join orders o on o.order_id = op.order_id
group by o.order_id, op.product_id, op.add_to_cart_order, op.reordered, o.user_id, o.order_number, o.order_dow, o.order_hour_of_day, o.days_since_prior_order;



select o.product_id, o.user_id into user_product
from order_products_prior_2 o join (select user_id from order_products_train_2 group by user_id) a on a.user_id = o.user_id
group by o.product_id, o.user_id;



select o.order_id, o.product_id, o.add_to_cart_order, o.reordered, o.user_id, o.order_number, o.order_dow, o.order_hour_of_day, o.days_since_prior_order, p.product_name, p.aisle, p.department, p.category into order_products_prior_3 from order_products_prior_2 o join products_full p on p.product_id = o.product_id group by o.order_id, o.product_id, o.add_to_cart_order, o.reordered, o.user_id, o.order_number, o.order_dow, o.order_hour_of_day, o.days_since_prior_order, p.product_name, p.aisle, p.department, p.category;


select o.order_id, o.product_id, o.add_to_cart_order, o.reordered, o.user_id, o.order_number, o.order_dow, o.order_hour_of_day, o.days_since_prior_order, p.product_name, p.aisle, p.department, p.category into order_products_train_3 from order_products_train_2 o join products_full p on p.product_id = o.product_id group by o.order_id, o.product_id, o.add_to_cart_order, o.reordered, o.user_id, o.order_number, o.order_dow, o.order_hour_of_day, o.days_since_prior_order, p.product_name, p.aisle, p.department, p.category;

SELECT * into orders_2 FROM orders WHERE eval_set <> 'test';

select user_id into user_sample from orders_2 WHERE eval_set = 'train' group by user_id order by random() limit 20000;

SELECT o.order_id, o.user_id, o.eval_set, o.order_number, o.order_dow, o.order_hour_of_day, o.days_since_prior_order into orders_sample
FROM orders_2 o JOIN user_sample u on u.user_id = o.user_id
GROUP BY o.order_id, o.user_id, o.eval_set, o.order_number, o.order_dow, o.order_hour_of_day, o.days_since_prior_order;

SELECT op.order_id, op.product_id, op.add_to_cart_order, op.reordered INTO order_products_prior_sample
FROM order_products_prior op JOIN orders_sample o on o.order_id = op.order_id
GROUP BY op.order_id, op.product_id, op.add_to_cart_order, op.reordered;

SELECT op.order_id, op.product_id, op.add_to_cart_order, op.reordered INTO order_products_train_sample
FROM order_products_train op JOIN orders_sample o on o.order_id = op.order_id
